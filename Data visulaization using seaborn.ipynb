### -loading libraries and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the samples diamond data
df = pd.read_csv("D:/diamonds2.csv")
df.head()
```

```python
df.tail()
```

### Explaining each part of diamond dataset

- **carat**: Weight of the diamond (0.2 – 5.1)
- **cut**: Quality of cut (Fair, Good, Very Good, Premium, Ideal)
- **color**: Diamond color from D (best) to J (worst)
- **clarity**: A measurement of how clear the diamond is
- **depth**: Total depth percentage
- **table**: Width of top of diamond relative to widest point (43 – 95)
- **price**: Price in USD (326 – 18,823)
- **x**: Length in mm
- **y**: Width in mm
- **z**: Depth in mm

## BASIC PLOT USING SEABORN

```python
#seaborn scatter plot
sns.scatterplot(x='carat',y='price',data=df)
plt.title("Seaborn Scatter plot of carat vs price")
plt.show()
```

```python
sns.scatterplot(x='carat',y='price',data=df,hue='cut')
plt.title("Seaborn Scatter plot of carat vs price")
plt.show()
```

```python
#seaborn line plot
#shaded area is standard deviation
sns.lineplot(x='carat',y='price',data=df)
plt.title("Seaborn Line plot of carat vs price")
plt.show()
```

```python
figure=plt.figure(figsize=(12,3))
sns.lineplot(x='carat',y='price',data=df,hue='cut')
plt.title("Seaborn Line plot of carat vs price")
plt.show()
```

```python
#individual lines for all cuts using facetgrid
g=sns.FacetGrid(df,col='cut',col_wrap=3,height=4)
g.map(sns.lineplot,'carat','price')
g.add_legend()
plt.show()
```

## INTERMEDIATE PLOT USING SEABORN

```python
#boxplot
sns.boxplot(x='cut',y='price',data=df)
plt.title("Box plot of price by cut")
plt.show()
```

```python
#violin plot
sns.violinplot(x='color',y='price',data=df)
plt.title("Violin plot of price by cut")
plt.show()
```

```python
#Count plot
sns.countplot(x='cut',data=df)
plt.title("Count plot of cut")
plt.show()
```

```python
sns.countplot(x='cut',data=df,hue='color')
plt.title("Count plot of cut")
plt.show()
```

## ADVANCED PLOT WITH SEABORN

```python
#heatmap
#build in function of corr in python
#0.5> and .5< are the significant data
#here depth and table are the lowest the larger the depth the table will be smaller of diamond
correlation=df[['carat','depth','table','price','x','y','z']].corr()
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title('Heatmap of Correlation matrix')
```

## ENHANCING PLOTS WITH SEABORN FOR REPORT READY VISUALIZATION

```python
#setting seaborn style
sns.set_style("whitegrid")#white,dark,darkgrid,ticks
sns.set_context("paper")#paper,notebook,talk,poster
#customize font size
sns.set_context("paper",font_scale=1.0)
#customize color palette
sns.set_palette("dark")#deep,muted,bright,pastel,dark,colorblind
#use set theme for all in one
#sns.set_theme(style='Whitegrid',palette='dark',context='paper',font_scale=1.0)
#grid line 
plt.grid(True,linestyle='--',alpha=0.5) 
#could also rotate axis for better reading
plt.xticks(rotation=360)#90,45,180
#we can also change scaling
#plt.yscale('log') #10^4

#example plot for seeing changes
sns.scatterplot(x='carat',y='price',data=df)
plt.title("Seaborn Scatter plot of carat vs price")
plt.show()
```

```python
#creating and using custom color palette
#number of color shoul be according to bars 
cus_palette=sns.color_palette(['#FF0000','#FFFF00','#06402B','#A020F0','#000000'])
sns.set_palette(cus_palette)

#example plot to see changes
sns.countplot(x='cut',data=df,hue='color')
plt.title("Count plot of custom color palette")
plt.show()
```

```python
figure=plt.figure(figsize=(12,3))
sns.boxplot(x='cut',y='price',data=df)
#customizing legend
plt.legend(title='Cut Types',labels=df['cut'].unique(),loc='upper left',bbox_to_anchor=(1,1),frameon=True)
#sample plot for changes
plt.title("Box plot of price by cut")
plt.show()
```
